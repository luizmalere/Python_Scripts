```python
df=pd.read_csv('http://bit.ly/2vmLyvU')
target = 'radius_mean'

max = df[target].max()
min = df[target].min()
range = max - min
print('The max is %.1f, the min is %.1f, the range is %.1f' %(max,min,range))
```

```python
mean = df[target].mean()
median = df[target].median()
mode = df[target].mode()
print('The mean is %.1f, the median is %.1f and the mode is %.1f.' % (mean, median, mode))
```

```python
q1 = df[target].quantile(0.25)
q3 = df[target].quantile(0.75)
iqr = q3-q1
print('The lower quartile is %.2f, the upper quartile is %.2f and the interquartile range is %.2f' % (q1, q3, iqr))
```
